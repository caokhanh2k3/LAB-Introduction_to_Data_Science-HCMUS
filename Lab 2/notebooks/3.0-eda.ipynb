```python
# YOUR CODE HERE
# Create a new dataframe
sum_total_population_df = pd.DataFrame()

# YOUR CODE HERE

# Find average for each country
sum_total_population_df["Sum.Total Population"] = cleaned_df.groupby("Country")["Total Population"].sum()

# Set the index as a column - countries
sum_total_population_df["Country"] = sum_total_population_df.index

# Drop the index
sum_total_population_df.reset_index(drop=True, inplace=True)

# Sort the rows based of Sum Total Population
sum_total_population_df.sort_values("Sum.Total Population", inplace=True, ascending=False)


top10_largest_sum_total_population_df = sum_total_population_df.head(10)

top10_largest_sum_total_population_df
```

```python
result = pd.DataFrame()
# Sá»­ dá»¥ng merge Ä‘á»ƒ káº¿t há»£p dá»¯ liá»‡u tá»« cleaned_df vÃ  top10_largest_sum_total_population_df dá»±a trÃªn cá»™t 'country'
result = cleaned_df.merge(top10_largest_sum_total_population_df, on='Country')

# In ra dá»¯ liá»‡u káº¿t quáº£
result

# create a copy
df_copy = result.copy()

# Exclude the categorical features from the matrix
df_copy.drop(['Year', 'Country', 'Sum.Total Population'], inplace=True, axis='columns')

# plot a correlation matrix
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df_copy.corr(), cmap='RdBu', center=0,ax=ax)
plt.show()
```

**YOUR ANSWER**
- When the conutry population is large. The features 'GDP in USD', 'Net income from Abroad', 'Agriculture value added (in USD)', 'Electric Power Consumption (kWh per capita)' will increase. As the population grows, it will boost the total income, income from abroad, and aid economic growth. In addition, a larger population requires ensuring domestic food sources, contributing to economic, energy, and agricultural growth.
- The features that could be dropped due to weak correlation are: 'Compulsory Education Dur.', 'National Income per Capita', 'Literacy rate'

## Making your own question 01? (1.5 points)

**YOUR ANSWER HERE**    
**A possible question is:** How does the correlation between features change in the top 10 countries with the HIGHEST GDP Per Capita?

**Answering this question will** we might see how features interaction with each other when the GDP Per Capita is large.

**How we answer this question**: we will consider the Average GDP Per Capita for countries across all the year. The GDP Per Capita = (GDP in USD / Total Population)

### Pre-processing

### Analyzing to get the answer

```python
# YOUR CODE HERE
# Create a copy of the DataFrame
new_cleaned_df = cleaned_df.copy()

new_cleaned_df['GDP Per Capita'] = new_cleaned_df['GDP in USD'] / new_cleaned_df['Total Population']

# Create a new dataframe
avg_GDP_Per_Capita_df = pd.DataFrame()

# YOUR CODE HERE

# Find average for each country
avg_GDP_Per_Capita_df["Avg.GDP Per Capita"] = new_cleaned_df.groupby("Country")["GDP Per Capita"].mean()

# Set the index as a column - countries
avg_GDP_Per_Capita_df["Country"] = avg_GDP_Per_Capita_df.index

# Drop the index
avg_GDP_Per_Capita_df.reset_index(drop=True, inplace=True)

avg_GDP_Per_Capita_df.sort_values("Avg.GDP Per Capita", inplace=True, ascending=False)


top10_largest_avg_GDP_Per_Capita_df = avg_GDP_Per_Capita_df.head(10)

top10_largest_avg_GDP_Per_Capita_df
```

```python
result = pd.DataFrame()
# Sá»­ dá»¥ng merge Ä‘á»ƒ káº¿t há»£p dá»¯ liá»‡u tá»« new_cleaned_df vÃ  top10_largest_avg_GDP_Per_Capita_df dá»±a trÃªn cá»™t 'country'
result = new_cleaned_df.merge(top10_largest_avg_GDP_Per_Capita_df, on='Country')

# In ra dá»¯ liá»‡u káº¿t quáº£
result

# create a copy
df_copy = result.copy()

# Exclude the categorical features from the matrix
df_copy.drop(['Year', 'Country', 'Avg.GDP Per Capita'], inplace=True, axis='columns')

# plot a correlation matrix
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df_copy.corr(), cmap='RdBu', center=0,ax=ax)
plt.show()
```

**Comment :**
- GDP Per Capita is an indicator that reflects the wealth of the population in a country. GDP Per Capita usually has a direct correlation with GDP in USD and an inverse relationship with Total Population.  
- From the chart, we can observe that as GDP Per Capita increases (indicating a wealthier population), the birth rate rises, and the death rate decreases. Employment in various sectors and the primary completion rate are likely to increase
